```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt

img = cv.imread('kucing.jpg', 0)
laplacian64f = cv.Laplacian(img,cv.CV_64F)
abs_laplacian64f = np.absolute(laplacian64f)
laplacian_8u = np.uint8(abs_laplacian64f)
sobelx64f = cv.Sobel(img,cv.CV_64F,1,0,ksize=3)
abs_sobelx64f = np.asbolute(sobelx64f)
sobelx_8u = np.uint8(abs_sobelx64f)
sobelx64f = cv.Sobel(img,cv.CV_64F,0,1,ksize=3)
abs_sobely64f = np.asbolute(sobelx64f)
sobely_8u = np.uint8(abs_sobelx64f)
magnitudesobel = cv.magnitude(sobelx64f, sobelx64f)
abs_sobel64f = np.asbolute(magnitudesobel)
sobel_8u = np.uint8(abs_sobelx64f)

plt.subplot(3,2,1), plt.imshow(img,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])

plt.subplot(3,2,2), plt.imshow(laplacian_8u,cmap = 'gray')
plt.title('Laplacian'), plt.xticks([]), plt.yticks([])

plt.subplot(3,2,3),plt.imshow(sobelx_8u,cmap = 'gray')
plt.title('Sobel x'), plt.xticks([]), plt.yticks([])

plt.subplot(3,2,4),plt.imshow(sobely_8u,cmap = 'gray')
plt.title('Sobel y'), plt.xticks([]), plt.yticks([])

plt.subplot(3,2,5),plt.imshow(sobel_8u,cmap = 'gray')
plt.title('Sobel'), plt.xticks([]), plt.yticks([])

cv.imshow(img)
cv.imshow(laplacian_8u)
cv.imshow(sobelx_8u)
cv.imshow(sobely_8u)
cv.imshow(sobel_8u)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python

import cv2
import numpy as np
import matplotlib.pyplot as plt

# Membaca citra
image = cv2.imread('daun.jpeg')

# Mengubah citra ke grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Menampilkan citra grayscale
plt.imshow(gray_image, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')
plt.show()

# Sobel Edge Detection
sobelx = cv2.Sobel(gray_image, cv2.CV_64F, 1, 0, ksize=3)
sobely = cv2.Sobel(gray_image, cv2.CV_64F, 0, 1, ksize=3)
magnitude = np.sqrt(sobelx**2 + sobely**2)
magnitude = cv2.convertScaleAbs(magnitude)

plt.figure(figsize=(10, 4))
plt.subplot(1, 3, 1)
plt.imshow(cv2.convertScaleAbs(sobelx), cmap='gray')
plt.title('Sobel - Horizontal (Gx)')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(cv2.convertScaleAbs(sobely), cmap='gray')
plt.title('Sobel - Vertical (Gy)')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(magnitude, cmap='gray')
plt.title('Sobel - Magnitude')
plt.axis('off')
plt.show()

# Laplacian Edge Detection
laplacian = cv2.Laplacian(gray_image, cv2.CV_64F)
laplacian = cv2.convertScaleAbs(laplacian)
plt.imshow(laplacian, cmap='gray')
plt.title('Laplacian Edge Detection')
plt.axis('off')
plt.show()

# Canny Edge Detection
canny_edges = cv2.Canny(gray_image, 100, 200)
plt.imshow(canny_edges, cmap='gray')
plt.title('Canny Edge Detection')
plt.axis('off')
plt.show()

# Image Segmentation
_, segmented_image = cv2.threshold(gray_image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
segmented_image = cv2.bitwise_not(segmented_image)
plt.imshow(segmented_image, cmap='gray')
plt.title('Segmented Image')
plt.axis('off')
plt.show()

# Apply mask to original image
mask = segmented_image == 255
segmented_daun = np.zeros_like(gray_image)
segmented_daun[mask] = gray_image[mask]

plt.imshow(segmented_daun, cmap='gray')
plt.title('Segmented Leaf with Original Intensities')
plt.axis('off')
plt.show()
```
